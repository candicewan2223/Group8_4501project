```python
# We import data and define our original dataset as df. As there will ba a warning due to the large dataset, we just use code to ignore it.

import pandas as pd
import warnings
warnings.filterwarnings('ignore')
df = pd.read_csv("311_Service_Requests_2020.csv")
df.head()
```

```python
# Define a subset of df as df1 where the zip code is 11101.

df1 = df[df["Incident Zip"] == 11101.0]
df1
```

```python
# Define a subset named df2 where zip code is 11101 and Complaint Type is Illegal Parking.

df2 = df1[df1["Complaint Type"] == "Illegal Parking"]


# Calculate the number of illegal parking incidents in 11101.

parking_counts_11101=df2['Complaint Type'].value_counts().astype(int)
parking_counts_11101 = parking_counts_11101.sum()
parking_counts_11101
```

```python
# Calculate the number of total incidents in 11101.

incidents_counts_11101=df1['Complaint Type'].value_counts().astype(int)
incidents_counts_11101 = incidents_counts_11101.sum()
incidents_counts_11101
```

```python
# Calculate the proportion of parking incidents in 11101.

proportion_11101 = parking_counts_11101/incidents_counts_11101
proportion_11101
```

```python
# Define a subset of df named df3 where Complaint Type is Illegal Parking.

df3 = df[df["Complaint Type"] == "Illegal Parking"]
df3
```

```python
# Calculate the number of illegal parking incidents in all zip codes.

parking_counts = df3['Complaint Type'].value_counts().astype(int)
parking_counts = parking_counts.sum()
parking_counts
```

```python
# Calculate the number of total incidents in all zip codes.

incidents_counts=df['Complaint Type'].value_counts().astype(int)
incidents_counts = incidents_counts.sum()
incidents_counts
```

```python
# Calculate the proportion of parking incidents in all zip codes.

global_proportion = parking_counts/incidents_counts
global_proportion
```

```python
# Check if illegal parking incidents are a larger fraction of incidents in 11101.

if proportion_11101 > global_proportion:
    higher_parking_proportion = True
else:
    higher_parking_proportion = False

higher_parking_proportion
```
