```python
import pandas as pd
import warnings
warnings.filterwarnings('ignore')
df = pd.read_csv("311_Service_Requests_2020.csv")
df.head()
```

```python
#zip code is 11101
zipcode=df[df["Incident Zip"]==11101]
zipcode
```

```python
#derive data in the year 2020
zipcode['year'] = pd.DatetimeIndex(zipcode['Created Date']).year
year2020=zipcode[zipcode["year"]==2020]
year2020
```

```python
#derive the complaint type column
complaint=year2020["Complaint Type"]
complaint
```

```python
#get the top 10 causes of calls to 311 at zipcode 11101
top10=complaint.value_counts().head(10)
top10
```
